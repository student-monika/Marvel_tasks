```python
# Importing necessary libraries
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd

# Load the Iris dataset
df = px.data.iris()

# Display the first few rows of the dataset
df.head()

# 1. Scatter Plot - Sepal Length vs Petal Length
fig_scatter = px.scatter(df,
                         x="sepal_length",
                         y="petal_length",
                         color="species",
                         title="Sepal Length vs Petal Length")
fig_scatter.show()

# 2. Histogram - Distribution of Sepal Width
fig_histogram = px.histogram(df,
                             x="sepal_width",
                             color="species",
                             title="Distribution of Sepal Width")
fig_histogram.show()

# 3. Box Plot - Petal Width by Species
fig_box = px.box(df,
                 y="petal_width",
                 color="species",
                 title="Box plot of Petal Width by Species")
fig_box.show()

# 4. Heatmap - Correlation between Features
# Remove non-numeric column (species) for correlation computation
df_numeric = df.drop(columns=['species'])

# Compute correlation matrix
df_corr = df_numeric.corr()

# Create the heatmap
fig_heatmap = px.imshow(df_corr,
                        text_auto=True,
                        title="Heatmap of Feature Correlations")
fig_heatmap.show()

# 5. Line Plot - Sepal Length vs Petal Length
fig_line = px.line(df,
                   x="sepal_length",
                   y="petal_length",
                   color="species",
                   title="Line Plot of Sepal Length vs Petal Length")
fig_line.show()
```
