```python
# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn import datasets

# Load Iris dataset
iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target

# Line Plot: Simple line plot for Sepal Length
plt.plot(df['sepal length (cm)'], label='Sepal Length')
plt.xlabel('Sample Index')
plt.ylabel('Sepal Length (cm)')
plt.title('Line Plot of Sepal Length')
plt.legend()  # Add a legend
plt.show()  # Show the plot

# Scatter Plot: Plot Sepal Length vs Sepal Width
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], c=df['species'], cmap='viridis')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Scatter Plot: Sepal Length vs Sepal Width')
plt.show()

# Bar Plot: Count of each species
species_count = df['species'].value_counts()
species_count.plot(kind='bar', color=['blue', 'green', 'red'])
plt.title('Bar Plot of Species Count')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()

# Histogram: Distribution of Sepal Length
df['sepal length (cm)'].plot(kind='hist', bins=20, color='purple', edgecolor='black')
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.show()

# Pie Chart: Distribution of species
species_count.plot(kind='pie', autopct='%1.1f%%', colors=['#ff9999', '#66b3ff', '#99ff99'], figsize=(6, 6))
plt.title('Pie Chart of Species Distribution')
plt.ylabel('')  # Remove the default label
plt.show()

# Box Plot: Distribution of features (Sepal length, width, etc.)
sns.boxplot(data=df.iloc[:, :-1])  # Exclude the 'species' column
plt.title('Box Plot of Iris Features')
plt.show()

# Heatmap: Correlation between features
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap of Feature Correlations')
plt.show()
```
